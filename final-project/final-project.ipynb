```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
```

```python
#top 5 entries
df = pd.read_csv('diabetes.csv')
df.head(5)
```

```python
#bottom 5 entries
df.tail(5)
```

```python
#handle missing values (data cleansing)
df.fillna(df.mean(), inplace=True)
df.info()
df.head(5)
```

```python
#pie chart
plt.figure(figsize=(10,5))
plt.pie(df['Outcome'].value_counts(), labels=['Healthy', "Diabetic"], colors=['#b6e1f2','#6890a1'], shadow=True, autopct='%1.1f%%')
plt.legend(title="Outcome")
plt.title('Count of Healthy vs Diabetic People', fontsize=20, color='#6869a1')
```

```python
#univariate analysis: histograms and boxplots
def plot_univariate(df):
    colors= ['blue','purple','darkgreen','lightgreen','blue','purple','darkgreen','lightgreen']
    fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 20))
    axes = axes.flatten()

    for i, col in enumerate(df.columns[:-1]):  # Exclude the 'Outcome' column
        sns.histplot(df[col], kde=True, ax=axes[i], color=colors[i])
        axes[i].set_title(f'Distribution of {col}')
    
    plt.tight_layout()
    plt.show()

    colors= ['lightblue','yellow','pink','lavender','lightblue','yellow','pink','lavender']
    fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 20))
    axes = axes.flatten()

    for i, col in enumerate(df.columns[:-1]):  # Exclude the 'Outcome' column
        sns.boxplot(y=df[col], ax=axes[i], color=colors[i])
        axes[i].set_title(f'Boxplot of {col}')
    
    plt.tight_layout()
    plt.show()

plot_univariate(df)
```

These univariate plots show a general distribution or the trend of the data in each columns before we attempt to analyse it and answer the question.

Question: 
What is a better indicator of diabetes among the Pima Indians: pathological factors (such as glucose and insulin levels) or family history (pedigree function)?

```python
#scatter plots
def plot_bivariate(df):
    sns.pairplot(df, hue='Outcome', palette='Spectral')
    plt.show()

plot_bivariate(df)

#boxplots
def plot_boxplots(df):
    fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 20))
    axes = axes.flatten()

    for i, col in enumerate(df.columns[:-1]):  # Exclude the 'Outcome' column
        sns.boxplot(x='Outcome', y=col, data=df, ax=axes[i], palette='bwr')
        axes[i].set_title(f'Boxplot of {col} by Outcome')
    
    plt.tight_layout()
    plt.show()

plot_boxplots(df)
```

These are the bivariate plots that show correlations between different factors.

The first plot shows a general overview of the correlation between data throguh scatter plots, out of which we will explore some in more detail to reach a conclusion.

The bivariate boxplots show the distribution of data by outcome.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

#3d ccatter plot for DiabetesPedigreeFunction, Glucose and Outcome
def plot_3d_scatter(df):
    fig = plt.figure(figsize=(15, 10))
    ax = fig.add_subplot(111, projection='3d')
    scatter = ax.scatter(df['DiabetesPedigreeFunction'], df['Glucose'], df['Insulin'], 
                         c=df['Outcome'], cmap='viridis', s=50)
    ax.set_xlabel('Diabetes Pedigree Function')
    ax.set_ylabel('Glucose')
    ax.set_zlabel('Insulin')
    plt.title('3D Scatter Plot of Diabetes Pedigree Function, Glucose, and Insulin')
    legend = ax.legend(*scatter.legend_elements(), title="Outcome")
    ax.add_artist(legend)
    plt.show()

plot_3d_scatter(df)

import plotly.express as px

fig1 = px.scatter_3d(df, x='DiabetesPedigreeFunction', y='Glucose', z='Insulin', color='Outcome')
fig1.update_layout(title='Scatterplot on Pedigree vs Glucose vs Insulin based on Outcome', scene = dict(
                    xaxis_title='Diabetes Pedigree Function',
                    yaxis_title='Glucose',
                    zaxis_title='Insulin'))
fig1.show()
```

Through this 3-d scatterplot comparing both the inherited and physiological characteristics of the patient, we see that:

A high diabetes pedigree function generally results in appearance of diabetes

High glucose can have an affect or result in the appearance of diabetes

Insulin has a weaker correlation with the outcome in comparison

```python
def plot_correlation_scatter(df):
    #scatter plot: DiabetesPedigreeFunction vs Glucose by Outcome
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='DiabetesPedigreeFunction', y='Glucose', hue='Outcome', palette='cubehelix', data=df)
    plt.title('Scatter Plot of Diabetes Pedigree Function vs Glucose by Outcome')
    plt.xlabel('Diabetes Pedigree Function')
    plt.ylabel('Glucose')
    plt.show()

plot_correlation_scatter(df)

#bubble plot
def plot_bubble(df):
    plt.figure(figsize=(12, 8))
    plt.scatter(df['DiabetesPedigreeFunction'], df['Glucose'], 
                s=df['Pregnancies']*10, c=df['Outcome'], cmap='copper', alpha=0.6, edgecolors="w", linewidth=0.5)
    plt.xlabel('Diabetes Pedigree Function')
    plt.ylabel('Glucose')
    plt.title('Bubble Plot: Diabetes Pedigree Function vs Glucose')
    plt.colorbar(label='Outcome')
    plt.show()

plot_bubble(df)
```

These scatter and bubble plots show that there is somewhat of a correlation between glucose and the outcome, with most of the 0's being clustered towards the bottom and the 1's on top. The correlation with the pedigree is alttile weaker and there a few anomalies but the general trend shows a higher chance with a higher pedigree function.

```python
#violin plot for DiabetesPedigreeFunction by Outcome
def plot_violin(df):
    plt.figure(figsize=(10, 5))
    sns.violinplot(x='Outcome', y='DiabetesPedigreeFunction', data=df, palette='coolwarm')
    plt.title('Violin Plot of Diabetes Pedigree Function by Outcome')
    plt.show()

    plt.figure(figsize=(10, 5))
    sns.violinplot(x='Outcome', y='Glucose', data=df, palette='magma')
    plt.title('Violin Plot of Glucose by Outcome')
    plt.show()

    plt.figure(figsize=(10, 5))
    sns.violinplot(x='Outcome', y='Insulin', data=df, palette='Accent')
    plt.title('Violin Plot of Insulin by Outcome')
    plt.show()

plot_violin(df)
```

In these violin plots we see that:

The diabetes pedigree function is more concentrated towards the higher values in 1 as compared to an outcome of 0

The higher values of glucose clearly result in more appearaces of diabetes

Both the 0 and 1 outcomes are concentrated towards the lower values, thus not giving a clear conclusion in insulin

```python
#kde plot for DiabetesPedigreeFunction
def plot_kde(df):
    plt.figure(figsize=(12, 8))
    sns.kdeplot(df[df['Outcome'] == 0]['DiabetesPedigreeFunction'], label='Outcome 0', shade=True, color='blue')
    sns.kdeplot(df[df['Outcome'] == 1]['DiabetesPedigreeFunction'], label='Outcome 1', shade=True, color='red')
    plt.title('KDE Plot of Diabetes Pedigree Function by Outcome')
    plt.xlabel('Diabetes Pedigree Function')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    plt.figure(figsize=(12, 8))
    sns.kdeplot(df[df['Outcome'] == 0]['Glucose'], label='Outcome 0', shade=True, color='purple')
    sns.kdeplot(df[df['Outcome'] == 1]['Glucose'], label='Outcome 1', shade=True, color='green')
    plt.title('KDE Plot of Glucose by Outcome')
    plt.xlabel('Glucose')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

plot_kde(df)
```

In both the KDE plots, we see that a higher pedigree or glucose has a higher probability of appearance in diabetes, although it is more pronounced in glucose.

```python
#joint plot for DiabetesPedigreeFunction and Glucose
def plot_joint(df):
    sns.jointplot(x='DiabetesPedigreeFunction', y='Glucose', hue='Outcome', data=df, palette='autumn')
    plt.show()

plot_joint(df)
```

```python
#correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='Spectral', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

This heatmap shows the correlation between each column and we see that glucose and the outcome have the highest relative correlation as 0.47, as seen in the joint plot as well, pedigree is lower at 0.17 and insulin is the lowest of the three at 0.13.

```python
#Q1
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('all-menu.csv')
#top 5 entries
df.head(5)
```

```python
#bottom 5 entries
df.tail(5)
```

```python
#Q2 removing missing values
df = df.fillna("0")
df.head(3)
```

```python
skip_arr= ['Beverage_category', 'Beverage', 'Beverage_prep', 'Calories', ' Sodium (mg)', ' Total Carbohydrates (g) ', 'Cholesterol (mg)', ' Dietary Fibre (g)', ' Sugars (g)']
for col in df.columns:
    if col in skip_arr:
        continue
    df[col]= df[col].str.replace('%','')
    df[col]= df[col].str.replace(' ','')
    df[col]= df[col].str.replace('varies','')
    df[col]= df[col].str.replace('Varies','')
    df[col] = df[col].astype(float)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
```

```python
#line chart
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(df['Calories'].head(50), df['Cholesterol (mg)'].head(50), marker='o', color='#d1bdf2', markeredgewidth='2.0', markeredgecolor='purple', markerfacecolor='lavender', linestyle= ':')
plt.xlabel('Total Calories')
plt.ylabel('Cholesterol (mg)')
plt.title('Calories vs Cholesterol Line Plot', fontsize=15, color='#9282e0')

plt.subplot(1,2,2)
plt.plot(df[' Total Carbohydrates (g) '].head(50), df[' Sugars (g)'].head(50), marker='o', color='#f2cbf0', markeredgewidth='2.0', markeredgecolor='pink', markerfacecolor='#fce6fb', linestyle= '--')
plt.xlabel('Total Carbohydrates')
plt.ylabel('Toal Sugars (g)')
plt.title('Carbohydrates vs Sugars Line Plot', fontsize=15, color='#b53ead')
```

```python
#bar plot
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.bar(x=df['Beverage_category'].unique(), height=df['Beverage_category'].value_counts().to_list(), color=['#b6e1f2', '#b6bff2', '#6890a1', '#6869a1'])
plt.xlabel('Beverage Categories')
plt.ylabel('Count')
plt.title('Count of Beverage Categories', fontsize=15, color='#30436b')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light'
)

plt.subplot(1,2,2)
plt.bar(x=df['Beverage_prep'].unique(), height=df['Beverage_prep'].value_counts().to_list(), color=['#bef7c5', '#779e65', '#8fdbb6', '#478064'])
plt.xlabel('Beverage Preparation')
plt.ylabel('Count')
plt.title('Count of Beverage Preparation', fontsize=15, color='#2f694d')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light'
)
plt.show()
```

```python
#area chart
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(df['Calories'].head(15), df['Cholesterol (mg)'].head(15), marker='o', color='orange', markeredgewidth='2.0', markeredgecolor='orange', markerfacecolor='#fce8d4', linestyle= ':')
plt.fill_between(df['Calories'].head(15), df['Cholesterol (mg)'].head(15), color='#ffdbba')
plt.xlabel('Total Calories')
plt.ylabel('Cholesterol (mg)')
plt.title('Calories vs Cholesterol Area Plot', fontsize=15, color='#d9873b')

plt.subplot(1,2,2)
plt.plot(df[' Total Carbohydrates (g) '].head(15), df[' Sugars (g)'].head(15), marker='o', color='red', markeredgewidth='2.0', markeredgecolor='red', markerfacecolor='#fac3c3', linestyle= '--')
plt.fill_between(df[' Total Carbohydrates (g) '].head(15), df[' Sugars (g)'].head(15), color='#fac3c3')
plt.xlabel('Total Carbohydrates')
plt.ylabel('Toal Sugars (g)')
plt.title('Carbohydrates vs Sugars Area Plot', fontsize=15, color='#cc5050')
```

```python
#dendogram
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
from scipy.spatial.distance import pdist
```

```python
df_plot= df.drop(columns=['Beverage_category', 'Beverage', 'Beverage_prep', 'Vitamin A (% DV) ', 'Vitamin C (% DV)', ' Calcium (% DV) ', 'Iron (% DV) ', 'Caffeine (mg)'], axis=1)

plt.figure(figsize=(10,5))
link = linkage(df_plot, 'ward')
dendrogram(link, color_threshold=500, truncate_mode='lastp', p=10)
plt.xticks(rotation=90)
plt.title('Dendrogram Plot', fontsize=20, color='orange')
plt.show()
```

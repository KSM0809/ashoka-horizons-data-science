```python
#Q1
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('nutrition-drinks.csv')
#top 5 entries
df.head(5)
```

```python
#bottom 5 entries
df.tail(5)
```

```python
#Q2 removing missing values
df = df.fillna("0")
df.head(3)
```

```python
df= df.rename(columns={'Unnamed: 0':'Name'})
skip_arr= ['Name']
for col in df.columns:
    if col in skip_arr:
        continue
    df[col]= df[col].str.replace('-','0')
    df[col] = df[col].astype(float)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
```

```python
#line chart
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(df['Calories'].head(50), df['Carb. (g)'].head(50), marker='o', color='#d1bdf2', markeredgewidth='2.0', markeredgecolor='purple', markerfacecolor='lavender', linestyle= ':')
plt.xlabel('Total Calories')
plt.ylabel('Carbohydrates (g)')
plt.title('Calories vs Carbohydrates Line Plot', fontsize=15, color='#9282e0')

plt.subplot(1,2,2)
plt.plot(df['Calories'].head(50), df['Fat (g)'].head(50), marker='o', color='#f2cbf0', markeredgewidth='2.0', markeredgecolor='pink', markerfacecolor='#fce6fb', linestyle= '--')
plt.xlabel('Total Calories')
plt.ylabel('Toal Fat (g)')
plt.title('Calories vs Fat Line Plot', fontsize=15, color='#b53ead')
```

```python
#bar plot
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.bar(x=df['Protein'].unique(), height=df['Protein'].value_counts().to_list(), color=['#b6e1f2', '#b6bff2', '#6890a1', '#6869a1'])
plt.xlabel('Protein (g)')
plt.ylabel('Count')
plt.title('Count of Protein', fontsize=15, color='#30436b')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light'
)

plt.subplot(1,2,2)
plt.bar(x=df['Fiber (g)'].unique(), height=df['Fiber (g)'].value_counts().to_list(), color=['#bef7c5', '#779e65', '#8fdbb6', '#478064'])
plt.xlabel('Fiber (g)')
plt.ylabel('Count')
plt.title('Count of Fiber', fontsize=15, color='#2f694d')
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light'
)
plt.show()
```

```python
#area chart
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(df['Calories'], df['Carb. (g)'], marker='o', color='orange', markeredgewidth='2.0', markeredgecolor='orange', markerfacecolor='#fce8d4', linestyle= ':')
plt.fill_between(df['Calories'], df['Carb. (g)'], color='#ffdbba')
plt.xlabel('Total Calories')
plt.ylabel('Carbohydrates (g)')
plt.title('Calories vs Carbohydrates Area Plot', fontsize=15, color='#d9873b')
plt.xticks(
    rotation=90, 
    horizontalalignment='right',
    fontweight='light'
)

plt.subplot(1,2,2)
plt.plot(df['Fat (g)'], df['Protein'], marker='o', color='red', markeredgewidth='2.0', markeredgecolor='red', markerfacecolor='#fac3c3', linestyle= '--')
plt.fill_between(df['Fat (g)'], df['Protein'], color='#fac3c3')
plt.xlabel('Total Fats (g)')
plt.ylabel('Protein (g)')
plt.title('Fats vs Protein Area Plot', fontsize=15, color='#cc5050')
```

```python
#boxplot
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(data=df['Calories'], color='#9282e0')
plt.xlabel('Calories')
plt.ylabel('Count')
plt.title('Count of Calories', fontsize=20, color='purple')
plt.subplot(1,2,2)
sns.boxplot(data=df['Protein'], color='#f2cbf0')
plt.xlabel('Protein (g)')
plt.ylabel('Count')
plt.title('Count of Protein', fontsize=20, color='pink')
plt.show()
```

```python
#countplot
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.countplot(x='Calories', data=df.reset_index(), palette=['#b6e1f2', '#b6bff2', '#6890a1', '#6869a1'])
plt.xlabel('Calories')
plt.ylabel('Count')
plt.title('Count of Calories', fontsize=18, color='#30436b')
plt.xticks(
    rotation=90, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-small' 
)
plt.subplot(1,2,2)
sns.countplot(x='Carb. (g)', data=df.reset_index(), palette=['#bef7c5', '#779e65', '#8fdbb6', '#478064'])
plt.xlabel('Total Carbohydrates (g)')
plt.ylabel('Count')
plt.title('Count of Total Carbohydrates', fontsize=18, color='#2f694d')
plt.xticks(
    rotation=90, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-small'  
)
plt.show()
```

```python
#scatterplot
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.scatterplot(df[['Calories', 'Fat (g)']], palette=['#d9873b', '#cc5050'])
plt.xlabel('Fat (g)')
plt.ylabel('Calories')
plt.title('Calories vs Fat', fontsize=20, color='maroon')
plt.subplot(1,2,2)
sns.scatterplot(df[['Carb. (g)', 'Protein']], palette=['#9282e0', '#f2cbf0'])
plt.xlabel('Protein (g)')
plt.ylabel('Total Carbohydrates (g)')
plt.title('Carbohydrates vs Protein', fontsize=20, color='purple')
plt.show()
```

```python
#3d scatterplot
import plotly.express as px
from mpl_toolkits.mplot3d import Axes3D
data = df[['Calories', 'Carb. (g)', 'Fat (g)']]
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Calories'], data['Fat (g)'], data['Carb. (g)'], color= 'skyblue')
ax.set_xlabel('Calories')
ax.set_ylabel('Total Carbohydrates (g)')
ax.set_zlabel('Fat (g)')
plt.show()
```

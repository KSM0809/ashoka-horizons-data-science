Columns 0, 1 and 2 contain nominal categorical data, as they contain qualitative data on the drink that isn't ordinal.

Columns 3, 7, 8, 9, 10, 11, 13, 14, 15, 16 and 17 are all discrete values. They contain integers with no intermediate values or decimals. THe last few columns contain percentage signs which is why they are read as objects. The last column showing the amount of caffeine has certain expcptional values that say 'varies'.

Columns 4, 5, 6 and 12 have continuous data as they contain float values that are intermediates and not distinct values.

```python
#Q3 find the null values in each column
df.isnull().sum()
```

```python
#making a new df without null values
del_array= np.where(df['Caffeine (mg)'].isnull())
del_row= np.array(del_array).astype(int).item()
df_new= df.drop(del_row)
df_new
```

```python
#Q4 rename 3 columns
df_new= df_new.rename(columns={'Beverage_prep':'Prep_type', 'Beverage':'Beverage_name', ' Sugars (g)':'Total Sugars (g)'})

#drop 2 columns
df_new= df_new.drop(columns=['Saturated Fat (g)', 'Trans Fat (g) ', ' Sodium (mg)'])
df_new
```

I chose to drop the trans fat and saturated fat columns as I thought such detail was not necessary since there is already a column on total fats for each drink.

I also dropped the column showing the sodium content for each drink as the values for minute and insignificant in milligrams.

```python
#Q5 find numerical columns
i=0
num_col= []
for c in df.columns:
    if df[df.columns[i]].dtypes== 'int64' or df[df.columns[i]].dtypes=='float64':
        num_col.append(df.columns[i])
    i= i+1
```

```python
# find minimum, maximum, mean, standard deviation, variance and sum of each column
i=0
for x in num_col:
    min= df[num_col[i]].min()
    print("The minimum value in", num_col[i], "is", min)
    max= df[num_col[i]].max()
    print("The maximum value in", num_col[i], "is", max)
    mean= df[num_col[i]].mean()
    print("The mean value in", num_col[i], "is", mean)
    std= df[num_col[i]].std()
    print("The standard deviation in", num_col[i], "is", std)
    var= df[num_col[i]].var()
    print("The variance in", num_col[i], "is", var)
    sum= df[num_col[i]].sum()
    print("The sum of the values in", num_col[i], "is", sum)
    print("\n")
    i+=1
```

```python
#Q6 randomly sample 25% 4 times in 4 subsets
sample_size = int(len(df_new) * .25)
df_sample1 = df_new.sample(sample_size)
df_new_remaining = df_new.drop(df_sample1.index)

df_sample2 = df_new_remaining.sample(sample_size)
df_new_remaining = df_new_remaining.drop(df_sample2.index)

df_sample3 = df_new_remaining.sample(sample_size)
df_new_remaining = df_new_remaining.drop(df_sample3.index)

df_sample4 = df_new_remaining
```
